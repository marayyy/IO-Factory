# Problem statement

The problem revolves around maximizing the profit for a landowner named Mr. X in Mars Land. He owns a strip of land and can choose to develop it with different types of properties: Theatres, Pubs, and Commercial Parks. Each property has a specific construction time and occupies a certain area of land. The goal is to determine the optimal mix of properties to develop within a given time period in order to maximize Mr. X's earnings. The earnings are based on the type of property developed and the duration it is operational. The challenge is to find the combination of properties that generates the highest possible profit within the specified time frame.

```python
asset_count = [0, 0, 0]  # No of assets (Pub, Theatre, Commercial Park)
time_unit = [4, 5, 10]  # Time units required to build each property

def entire_earnings(n):
    profit = 0
    profit_gained = [0, 0, 0]  # Profit for each property

    if n < 4:
        return 0  # assets which can not be developed with less than 4 time units

    # Calculate profit based on remaining time
    remaining_time = n - 4
    profit_gained[0] = remaining_time * 1000
    if n >= 5:
        remaining_time = n - 5
        profit_gained[1] = remaining_time * 1500
        if n >= 10:
            remaining_time = n - 10
            profit_gained[2] = remaining_time * 3000

    max_index = profit_gained.index(max(profit_gained))
    profit += profit_gained[max_index]
    asset_count[max_index] += 1

    return profit + entire_earnings(n - time_unit[max_index])  # Calculate earnings for remaining time units


n = int(input())
earnings = entire_earnings(n)

print('Time Unit:',n)
print("Earnings: $",earnings)
print("Solutions: T:", asset_count[1],"P:",asset_count[0],"C:",asset_count[2])
```
